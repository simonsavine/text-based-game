```python
from google.colab import files

uploaded = files.upload()

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import math


# Part 1, user input:

flag = True
while flag:
  try:
    filename = input("Input your image file name: ")
    img = Image.open(filename)
    width, height = img.size
    plt.imshow(img)
    flag = False

  except FileNotFoundError:
        flag = True

# Part 2, accumulate RGB values of image in pixel matrix:

width = 200
aspect_ratio = img.height / img.width
height = int(width * aspect_ratio * 0.45)
img = img.resize((width, height))

width, height = img.size

pixel_matrix = np.array(img)

# Part 3, convert RGB to single brightness numbers using mean:

RGB_arr = 0.21 * pixel_matrix[:, : , 0] + 0.72 * pixel_matrix[:, : , 1] + 0.07 * pixel_matrix[:, : , 2]

# Part 4, convert brightness to ASCII:

# Uncomment for more character nuances:

#characters = "`^\",:;Il!i~+_-?][}{1)(|\\/tfjrxnuvczXYUJCLQ0OZmwqpdbkhao*#MW&8%B@$"
characters = " .:-=+*#%@"
characters = characters[::-1]

ASCII_arr = []

for rgbval in RGB_arr.flatten():
  charNum = math.floor(rgbval / 255 * len(characters))
  ASCII_arr.append(characters[charNum])

# Part 5, print art!
print("Here is your image in beautiful ASCII art:")

for row in range(0, len(ASCII_arr), width):
  line = ASCII_arr[row:row+width]
  print("".join(line))
```
