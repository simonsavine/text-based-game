```python
import random

def rps():

  player_input = str.lower(input("Player 1, input your choice: "))
  computer_choice = random.choice(['rock','paper','scissors'])

  choices = {'rock': 1, 'paper': 2, 'scissors': 3}

  player_input_num = choices[player_input]
  computer_choice_num = choices[computer_choice]

  # Maps to 1, Player wins, maps to 2, Computer wins.
  outcomes = {
      (1, 2) : 2,
      (1, 3) : 1,
      (2, 1) : 1,
      (2, 3) : 2,
      (3, 1) : 2,
      (3, 2) : 1
  }

  # Case of a draw
  if player_input_num == computer_choice_num:
    print("It's a draw!")
    return

  # Player wins!
  if outcomes[player_input_num, computer_choice_num] == 1:
    print(player_input, "beats", computer_choice, "and Player wins!")

  # Computer wins!
  if outcomes[player_input_num, computer_choice_num] == 2:
    print(computer_choice, "beats", player_input, "and Computer wins!")
```

```python
rps()
```
